```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/kc_house_data.csv', sep = ',', encoding= 'latin-1')
df.head()
```

image.png

```python
df.info()
```

```python
# 1.2
plt.hist(df['price'])
plt.xlabel('Количество объектов спален')
plt.ylabel('тыс. руб.')
plt.title('Цена квартир');
```

```python
# 1.3
plt.hist(df['sqft_living'])
plt.xlabel('Количество объектов')
plt.ylabel('метр')
plt.title('Распределение квадратуры жилой площади');
```

```python
# 1.4
plt.hist(df['yr_built'])
plt.xlabel('год постройки')
plt.ylabel('количество объектов')
plt.title('Распределение объектов относительно года постройки');
```

```python
# 2.1
data = df['waterfront'].value_counts()
data
```

```python
plt.pie(data.values, labels=data.index, autopct='%1.1f%%')
plt.legend(data.index);
```

```python
# 2.2
data = df['floors'].value_counts()
plt.bar(data.index, data.values)
plt.xticks(rotation=30);
plt.grid()
```

```python
# 2.3
data = df['grade'].value_counts()
plt.bar(data.index, data.values)
plt.xticks(rotation=30);
plt.grid()
```
